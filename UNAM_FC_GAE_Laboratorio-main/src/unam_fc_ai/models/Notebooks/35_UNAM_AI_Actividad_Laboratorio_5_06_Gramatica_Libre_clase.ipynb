# Gramáticas Libres de Contexto (CFG) y las Probabilísticas (PCFG)

```python
import nltk
```

```python
from nltk import CFG
```

Se define la siguiente grámatica.

```python
grammar = CFG.fromstring("""
S -> NP VP
PP -> P NP
NP -> Det N | NP PP
VP -> V NP | VP PP
Det -> 'a' | 'the'
N -> 'dog' | 'cat'
V -> 'chased' | 'sat'
P -> 'on' | 'in'
 """)
```

```python
grammar
```

```python
grammar.start()
```

```python
grammar.productions()
```

```python
from nltk import PCFG
```

```python
toy_pcfg1 = PCFG.fromstring("""
S -> NP VP [1.0]
NP -> Det N [0.5] | NP PP [0.25] | 'John' [0.1] | 'I' [0.15]
Det -> 'the' [0.8] | 'my' [0.2]
N -> 'man' [0.5] | 'telescope' [0.5]
VP -> VP PP [0.1] | V NP [0.7] | V [0.2]
V -> 'ate' [0.35] | 'saw' [0.65]
PP -> P NP [1.0]
P -> 'with' [0.61] | 'under' [0.39]
""")
```

```python
g = CFG.fromstring("VP^<TOP> -> VBP NP^<VP-TOP>")
g.productions()[0].lhs()
```

# 1. Generar y visualizar el árbol de sintaxis

Lo más potente de NLTK es ver cómo la gramática "entiende" una oración. Puedes usar un RecursiveDescentParser para procesar una frase que cumpla con tu primera gramática.

```python

from nltk.parse import ChartParser

# Usamos ChartParser en lugar de RecursiveDescentParser
chart_parser = ChartParser(grammar)

sent = "the dog chased a cat".split()


for tree in chart_parser.parse(sent):
    print(tree)
    # tree.draw()  # Descomenta esto si estás en local para abrir una ventana con el árbol
```

# 2. Manejo de Ambigüedad (El "Porqué" de las PCFG)
Tu segunda gramática (toy_pcfg1) incluye probabilidades. Esto se usa para resolver la ambigüedad. Puedes añadir este ejemplo para mostrar cómo una oración puede tener dos sentidos, pero uno es más probable que el otro.

```python
from nltk.parse import viterbi

# Una oración ambigua clásica: "I saw the man with a telescope"
sent_pcfg = "I saw the man".split() # Ajusta según las palabras exactas de tu toy_pcfg1

# Usamos el algoritmo de Viterbi para encontrar el árbol más probable
viterbi_parser = viterbi.ViterbiParser(toy_pcfg1)

for tree in viterbi_parser.parse(sent_pcfg):
    print(tree)
```


# 3. Conceptos clave para tu explicación
Si estás armando una presentación o un notebook de clase, estos puntos te ayudarán a explicar qué está pasando en la imagen que subiste:

- Símbolos No Terminales: Son las etiquetas grandes como S (Sentence), NP (Noun Phrase) o VP (Verb Phrase). Son "variables" que se deben expandir.

- Símbolos Terminales: Son las palabras finales (ej. 'dog', 'chased'). Aquí termina la expansión.

- Reglas de Producción: El formato A -> B C significa que el símbolo A puede ser reemplazado por B y C.

- Probabilidades (PCFG): Explica que los números al lado de las reglas (como [0.5]) indican qué tan común es esa estructura. La suma de las probabilidades de todas las expansiones de un mismo símbolo (ej. todas las NP) debe ser 1.0.

# 4. Un toque extra: Generación Aleatoria
Para demostrar que la gramática "sabe" el idioma, puedes hacer que genere oraciones aleatorias que sigan esas reglas:

```python
from nltk.parse.generate import generate

# Genera las primeras 5 oraciones posibles según tu gramática
for sentence in generate(grammar, n=5):
    print(' '.join(sentence))
```
