```python
img1 = cv.imread('pythonNLP.jpg')
plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
```

```python
gauss = cv.GaussianBlur(img1,(7,7),0)
unsharp_img1 = cv.addWeighted(img1, 2, gauss,-1,0)
```

```python
plt.imshow(cv.cvtColor(unsharp_img1,cv.COLORMAP_OCEAN),cmap='gray')
```

```python
img1 = cv.imread('pythonNLP.jpg')
```

```python
img= cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(img, cmap='gray')
```

```python
ret, thresh1 = cv.threshold(img,120,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
reth,tresh3 = cv.threshold(img,127,255,cv.THRESH_TRUNC)
ret,thresh4 = cv.threshold(img,127,255,cv.THRESH_TOZERO)
ret,thresh5 = cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)
```

```python
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, tresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'ocean',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
