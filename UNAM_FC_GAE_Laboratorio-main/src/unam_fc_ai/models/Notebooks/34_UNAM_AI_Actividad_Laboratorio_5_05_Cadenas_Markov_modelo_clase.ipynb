```python
import nltk
from nltk.corpus import treebank
```

```python
from nltk.tag import hmm
```

En la figura 2 se muestran los pasos iniciales para entrenar un etiquetador POS con el modelo oculto de Markov. Se utiliza el corpus treebank de NLTK y se generan dos conjuntos de datos, el de entrenamiento con 3000 datos y el de pruebas con 914.

```python
nltk.download('treebank')
```

```python
train_data = treebank.tagged_sents()[:3000]
```

```python
test_data =treebank.tagged_sents()[3000:]
```

```python
print('Tamaño del conjunto de entrenamiento: ', len(train_data))
```

```python
print('Tamaño del conjunto de pruebas: ', len(test_data))
```

```python
print('Ejemplo de conjunto de entrenamiento: ',train_data[0])
```

En la figura 3 se observan las instrucciones para configurar y entrenar al modelo con los datos de entrenamiento previamente etiquetados. El modelo generado tiene 46 estados. Nota que después de entrenar el modelo, se realiza el etiquetado para ciertas oraciones en inglés que son previamente tokenizadas con la función split().

```python
trainer = hmm.HiddenMarkovModelTrainer()
```

```python
tagger = trainer.train_supervised(train_data)
```

```python
print(tagger)
```

```python
"Today is a good day .".split()
```

```python
print(tagger.tag("Today is a good day .".split()))
```

```python
print(tagger.tag("Today is a good day .".split()))
```

Al evaluar el modelo obtenido con los datos de prueba, se observa un rendimiento del 36.84%, sin embargo, en los datos de entrenamiento hay un rendimiento del 98.16%, como se observa en la figura 4.

```python
print(tagger.tag("Joe met Joanne in New Dalhi.".split()))
```

Estos niveles de error en los datos de prueba sugieren que el modelo se sobreentrenó, que los datos de entrenamiento son insuficientes para cubrir los casos del conjunto de pruebas o que el proceso de tokenización de las oraciones no fue suficiente, por lo que es posible utilizar normalización, lematización o stemming para mejorar la predicción. En la figura 5 se muestra que hay un 35% de palabras del conjunto de pruebas que no existen en el conjunto de entrenamiento.

En la figura 6 se puede ver que al aplicar stemming, la extracción de la raíz de una palabra previo al entrenamiento mejora el rendimiento del modelo en 36.84
%. Para lograr que el modelo generalice bien, es probable que se requiera un mayor preprocesamiento de los datos.

```python
print(tagger.evaluate(test_data))
```
