1. La Estructura: El defaultdict anidado
La línea model = defaultdict(lambda: defaultdict(lambda: 0)) es la base.

Se le llama modelo porque es el contenedor que "aprenderá" las reglas del lenguaje.

En este caso, está diseñado para guardar: {(palabra1, palabra2): {palabra3: probabilidad}}.

2. La fase de "Entrenamiento": Conteos
En el primer bloque for, el código recorre el corpus (textos de Reuters) buscando grupos de tres palabras (trigramas).

¿Qué hace? Cuenta cuántas veces la palabra w3 aparece justo después del par (w1, w2).

En este punto, el diccionario es solo una tabla de frecuencias. Aún no es un modelo estadístico útil, solo una base de datos de conteos.

3. La fase de "Cálculo": De conteos a probabilidades
Aquí es donde ocurre la magia. El segundo bloque for transforma esos números en probabilidades:

Toma el total de veces que ocurrió el par (w1, w2).

Divide el conteo de w3 entre ese total.

Resultado: Ahora el modelo no dice "la secuencia 'El gato come' apareció 5 veces", sino "Si veo 'El gato', hay un 80% de probabilidad de que la siguiente palabra sea 'come'".

¿A qué se le llama "Modelo" entonces?
En Machine Learning y NLP, el modelo es el resultado final de este proceso: el conjunto de parámetros (las probabilidades) que permiten hacer predicciones.

```python
#Variables para el modelo
model = defaultdict(lambda: defaultdict(lambda: 0))

#Conteo de frecuencia de co-ocurrencias
for sentence in reuters.sents():
    for w1, w2, w3 in trigrams(sentence, pad_right=True, pad_left=True):
        model[(w1, w2)][w3] += 1

# Convertir el conteo en probabilidades
for w1_w2 in model:
    total_count = float(sum(model[w1_w2].values()))
    for w3 in model[w1_w2]:
        model[w1_w2][w3] /= total_count
```

```python
import nltk
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.lm import MLE
```

```python
train_sentences = ['una manzana','una pera']
tokenized_text = [list(map(str.lower, nltk.tokenize.word_tokenize(sent)))
                   for sent in train_sentences]

n = 1
train_data, padded_vocab = padded_everygram_pipeline(n, tokenized_text)
model = MLE(n)
model.fit(train_data, padded_vocab)

test_sentences = ['una manzana', 'un gusano','manzana','La manzana roja es de mi casa junto con la pera','apple']
tokenized_text = [list(map(str.lower, nltk.tokenize.word_tokenize(sent)))
                  for sent in test_sentences]

test_Data,_= padded_everygram_pipeline(n,tokenized_text)
for test in test_Data:
  print("MLE Estima",[((ngram[-1],ngram[:-1]), model.score(ngram[-1],ngram[:-1])) for ngram in test])

test_data,_ = padded_everygram_pipeline(n, tokenized_text)

for i, test in enumerate(test_data):
  print("PP({0}):{1}".format(test_sentences[i],model.perplexity(test)))
```
