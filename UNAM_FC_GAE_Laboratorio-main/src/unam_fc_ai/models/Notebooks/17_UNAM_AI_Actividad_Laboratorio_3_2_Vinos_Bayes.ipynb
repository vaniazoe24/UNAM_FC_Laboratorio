## Análisis del ejercicio 6

En este ejercicio se utilizó un clasificador `RandomForestClassifier` para comparar su desempeño con el modelo Naive Bayes utilizado anteriormente.

Random Forest es un algoritmo de aprendizaje supervisado que construye varios árboles de decisión y combina sus resultados para realizar una predicción final. En este caso se utilizaron 200 árboles mediante el parámetro `n_estimators=200`.

También se midió el tiempo de entrenamiento y el tiempo de predicción usando la función `time.time()`. Esto permite observar empíricamente cuánto tarda el algoritmo en ejecutarse sobre el conjunto de datos.

### Complejidad computacional

Sea:

- `n` el número de muestras de entrenamiento.
- `m` el número de características.
- `t` el número de árboles del bosque.
- `d` la profundidad promedio de cada árbol.

### Complejidad de tiempo

El entrenamiento de un Random Forest depende del número de árboles y del costo de construir cada árbol de decisión.

De forma aproximada, la complejidad de entrenamiento es:

\[
O(t \cdot n \cdot m \cdot \log n)
\]

donde:

- `t` es el número de árboles.
- `n` es el número de datos de entrenamiento.
- `m` es el número de características.
- `log n` aparece porque los árboles dividen recursivamente los datos.

Para la predicción, cada dato debe pasar por cada árbol hasta llegar a una hoja. Por eso, la complejidad aproximada de predicción es:

\[
O(t \cdot d)
\]

para una sola muestra, y para varias muestras de prueba sería:

\[
O(k \cdot t \cdot d)
\]

donde `k` es el número de muestras de prueba.

### Complejidad de espacio

La complejidad de espacio depende de la cantidad de árboles almacenados y del tamaño de cada árbol.

De forma aproximada, la complejidad espacial es:

\[
O(t \cdot n)
\]

ya que cada árbol puede almacenar nodos generados a partir de las muestras de entrenamiento.

El clasificador Random Forest obtuvo un desempeño comparable al de Naive Bayes. Aunque en esta partición no mejora la exactitud porque Naive Bayes ya obtuvo el valor máximo de 1.0, Random Forest sigue siendo una alternativa adecuada para comparar modelos supervisados.

Además, al medir el tiempo de entrenamiento y predicción, se observa que Random Forest suele ser más costoso computacionalmente que Naive Bayes, debido a que construye múltiples árboles de decisión. Sin embargo, puede ser más robusto en conjuntos de datos más complejos.

La complejidad de tiempo del entrenamiento de Random Forest es aproximadamente \(O(t \cdot n \cdot m \cdot \log n)\), donde `t` es el número de árboles, `n` el número de muestras y `m` el número de características. La complejidad de predicción es aproximadamente \(O(k \cdot t \cdot d)\), donde `k` es el número de muestras de prueba y `d` la profundidad promedio de los árboles. La complejidad de espacio es aproximadamente \(O(t \cdot n)\), debido al almacenamiento de los árboles generados.
