En el ejemplo de la figura 4 se presenta una aplicación de las etiquetas POS, identificar los verbos más comunes en un corpus. Se usa el corpus Brown.

```python
from nltk.inference.tableau import Categories
nltk.download('brown')
from nltk.corpus import brown
noticias_etiq_brown = brown.tagged_words(categories='adventure',tagset = 'universal')
```

```python
import nltk
nltk.download('universal_tagset')
```

```python
word_tag_fd = nltk.FreqDist(noticias_etiq_brown)
```

```python
help("universal_tagset")
```

```python
[wt[0] for (wt,_) in word_tag_fd.most_common() if wt[1]=='VERB']
```
