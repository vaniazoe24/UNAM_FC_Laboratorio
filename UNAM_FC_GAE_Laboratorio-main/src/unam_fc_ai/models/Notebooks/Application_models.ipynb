```python
import sys
import os

# 1. Detectamos la ubicación actual del notebook
current_path = os.getcwd()

# 2. Subimos un nivel para llegar a la raíz y luego entramos a 'src'
# Esto funciona tanto en VS Code local como en Google Colab
project_root = os.path.abspath(os.path.join(current_path, '..'))
src_path = os.path.join(project_root, 'src')

# 3. Agregamos 'src' al buscador de Python (sys.path)
if src_path not in sys.path:
    sys.path.append(src_path)
    print(f"✅ Ambiente DEV configurado. Buscando librería en: {src_path}")

# 4. Ahora sí, la importación funcionará perfectamente
try:
    from unam_fc_ai.models.regression import linear_regression
    print("🚀 Librería unam_fc_ai importada con éxito.")
except ImportError as e:
    print(f"❌ Error de importación: {e}")


# Definición de datos de prueba (DEV)
# Ejemplo: Relación entre horas de estudio (X) y calificación (Y)
horas_estudio = [1, 2, 3, 4, 5, 6, 8, 10]
calificaciones = [5, 5.5, 6, 6.8, 7, 7.5, 8.8, 9.5]

print("--- Iniciando Cálculo de Regresión Lineal OLS ---")

# Invocamos la función de nuestro motor 'unam_fc_ai'
# Nota: Al ejecutar esto, verás el reporte de tiempo y memoria en la salida
intercepto_a, pendiente_b = linear_regression(horas_estudio, calificaciones)

print(f"\n✅ Modelo Entrenado:")
print(f"La ecuación es: y = {intercepto_a:.4f} + {pendiente_b:.4f}x")

# Ejemplo de predicción
nueva_hora = 7
prediccion = intercepto_a + (pendiente_b * nueva_hora)
print(f"\n💡 Predicción: Si estudias {nueva_hora} horas, tu calificación estimada es {prediccion:.2f}")




import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(horas_estudio, calificaciones, color='#003b5c', label='Datos Reales')
y_linea = [intercepto_a + pendiente_b * x for x in horas_estudio]
plt.plot(horas_estudio, y_linea, color='#d4af37', label='Línea de Regresión')
plt.title("Relación Horas vs Calificación (UNAM AI Lab)")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificación")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
