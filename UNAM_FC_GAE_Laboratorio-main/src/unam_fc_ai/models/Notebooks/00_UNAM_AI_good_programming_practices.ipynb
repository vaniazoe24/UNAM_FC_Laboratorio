````python
from typing import List, Tuple

def regresion_lineal(X: List[float], Y: List[float]) -> Tuple[float, float]:
    """
    Calcula los coeficientes de la recta de regresión lineal (y = a + bx).

    Args:
        X: Lista de números (float o int) que representan la variable independiente.
        Y: Lista de números (float o int) que representan la variable dependiente.

    Returns:
        Una tupla (a, b) donde 'a' es el intercepto y 'b' es la pendiente.

    Complexity:
        Tiempo: O(n) | Espacio: O(1)
    """
    if len(X) != len(Y):
        raise Exception("Los dos arreglos deben contener la misma cantidad de elementos")

    n: int = len(X)
    sum_x: float = 0
    sum_y: float = 0
    sum_x2: float = 0
    sum_xy: float = 0

    for i in range(n):
        sum_x += X[i]
        sum_y += Y[i]
        sum_x2 += X[i] ** 2
        sum_xy += X[i] * Y[i]

    # Fórmula de la pendiente: b = [nΣxy - (Σx)(Σy)] / [nΣx² - (Σx)²]
    denominador_b: float = (n * sum_x2) - (sum_x ** 2)
    if denominador_b == 0:
        raise ZeroDivisionError("La varianza de X es cero; no se puede calcular la regresión.")

    b: float = ((n * sum_xy) - (sum_x * sum_y)) / denominador_b

    x_prom: float = sum_x / n
    y_prom: float = sum_y / n
    a: float = y_prom - b * x_prom

    return a, b


def estimar_peso(a: float, b: float, estatura: float) -> float:
    """
    Calcula la estimación de peso para una sola estatura.

    Args:
        a: Intercepto calculado previamente.
        b: Pendiente calculada previamente.
        estatura: Valor numérico de la estatura a evaluar.

    Returns:
        El peso estimado como un valor flotante.
    """
    return a + b * estatura


def estimar_peso_masivo(a: float, b: float, list_estatura: List[float]) -> List[float]:
    """
    Calcula estimaciones para una lista múltiple de estaturas.

    Args:
        a: Intercepto.
        b: Pendiente.
        list_estatura: Lista de valores numéricos de estaturas.

    Returns:
        Lista de floats con los pesos estimados.
    """
    return [a + b * x for x in list_estatura]
````

**Validación de seguridad**: Se puede añadir una pequeña verificación para evitar la división por cero (denominador_b == 0), lo cual sucede si todos los valores de $X$ son idénticos.

**Consistencia**: Ahora cualquier persona que use la función sabrá que debe pasar listas de números y que recibirá una tupla o un flotante.

# Longitud de código

En programación, existe una convención llamada **PEP 8** que sugiere que las líneas de código no deberían exceder los 79 u 80 caracteres. Esto evita que tengas que hacer "scroll" horizontal y mejora mucho la legibilidad, especialmente en notebooks como Google Colab.

Para ajustar el texto sin perder información, aplicaremos saltos de línea estratégicos y aprovecharemos que Python permite romper líneas dentro de paréntesis, corchetes o llaves.

Aquí tienes el código ajustado para que sea "angosto" y cumpla con esos estándares:

````python
from typing import List, Tuple

# Ejercicio 5: Regresión lineal

def regresion_lineal(X: List[float], Y: List[float]) -> Tuple[float, float]:
    """
    Calcula los coeficientes de la recta de regresión
    lineal (y = a + bx).

    Args:
        X: Lista de números para la variable independiente.
        Y: Lista de números para la variable dependiente.

    Returns:
        Tupla (a, b) con el intercepto y la pendiente.

    Complexity:
        Tiempo: O(n) | Espacio: O(1)
    """
    if len(X) != len(Y):
        raise Exception(
            "Los arreglos deben tener la misma cantidad de elementos"
        )

    n: int = len(X)
    sum_x, sum_y, sum_x2, sum_xy = 0.0, 0.0, 0.0, 0.0

    for i in range(n):
        sum_x += X[i]
        sum_y += Y[i]
        sum_x2 += X[i] ** 2
        sum_xy += X[i] * Y[i]

    # Cálculo de la pendiente (b) con salto de línea
    denominador_b = (n * sum_x2) - (sum_x ** 2)

    if denominador_b == 0:
        raise ZeroDivisionError("Varianza de X es cero.")

    b = ((n * sum_xy) - (sum_x * sum_y)) / denominador_b

    # Cálculo del intercepto (a)
    a = (sum_y / n) - b * (sum_x / n)

    return a, b


def estimar_peso(a: float, b: float, estatura: float) -> float:
    """Calcula estimación simple de peso."""
    return a + b * estatura


def estimar_peso_masivo(
    a: float,
    b: float,
    list_estatura: List[float]
) -> List[float]:
    """Calcula estimaciones para múltiples estaturas."""
    return [a + b * x for x in list_estatura]
````

# Uso de logger para los log


````python
import logging
from typing import List, Tuple

# Configuración básica del logger
# En una biblioteca real, el usuario configuraría esto.
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger("IA_Library")


def regresion_lineal(X: List[float], Y: List[float]) -> Tuple[float, float]:
    """
    Calcula los coeficientes de la recta de regresión lineal (y = a + bx).

    Utiliza el método de mínimos cuadrados para encontrar la pendiente (b)
    y la intersección (a) que mejor se ajustan a los datos.

    Argumentos:
        X (list): Lista de variables independientes (ej. estatura).
        Y (list): Lista de variables dependientes (ej. peso).

    Retorna:
        tuple: (a, b) donde 'a' es la ordenada al origen y 'b' es la pendiente.

    Complejidad:
        Tiempo: O(n) - Recorre las listas una vez.
        Espacio: O(1) - Solo utiliza variables escalares para las sumatorias.
    """
    logger.info("Iniciando cálculo de regresión lineal.")

    try:
        if len(X) != len(Y):
            raise ValueError("Los arreglos X y Y deben tener la misma longitud.")

        n = len(X)
        if n == 0:
            raise ValueError("Las listas de entrada no pueden estar vacías.")

        sum_x = sum(X)
        sum_y = sum(Y)
        sum_x2 = sum(x**2 for x in X)
        sum_xy = sum(x*y for x, y in zip(X, Y))

        denominador = (n * sum_x2) - (sum_x ** 2)

        if denominador == 0:
            raise ZeroDivisionError("Varianza en X es cero (puntos verticales).")

        # Cálculo de la pendiente (b)
        # Fórmula: b = [n*Σ(xy) - Σx*Σy] / [n*Σ(x²) - (Σx)²]
        b = ((n * sum_xy) - (sum_x * sum_y)) / denominador

        print("--- Proceso de Cálculo ---")
        print(f"b = ({n}*{sum_xy} - {sum_x}*{sum_y}) / ({n}*{sum_x2} - ({sum_x}^2))")
        print(f"b = {b}")

        a = (sum_y / n) - b * (sum_x / n)

        print(f"a = {Y} - {b}*{X}")
        print(f"a = {a}")
        print("--------------------------")

        logger.info(f"Modelo entrenado exitosamente: a={a:.4f}, b={b:.4f}")
        return a, b

    except ValueError as ve:
        logger.error(f"Error de validación: {ve}")
        raise
    except ZeroDivisionError as zde:
        logger.error(f"Error matemático: {zde}")
        raise
    except Exception as e:
        logger.critical(f"Error inesperado en ejercicio5: {e}")
        raise




def estimar_peso(a: float, b: float, estatura: float) -> float:
    """Predice un valor con manejo de excepciones de tipo.
        Calcula y = a + bx.

        >>> estimar_peso(10, 2, 70)
        150
    """
    try:
        return a + b * estatura
    except TypeError:
        logger.error("Los parámetros a, b y estatura deben ser numéricos.")
        raise

def estimar_peso_masivo(a: float,
                        b: float,
                        list_estatura: List[float]) -> List[float]:
    """
    Realiza predicciones para una lista completa de valores.

    Argumentos:
        a (float): Intersección con el eje Y.
        b (float): Pendiente de la recta.
        list_estatura (list): Lista de valores X.

    Retorna:
        list: Lista con las estimaciones calculadas.
    """
    # Tiempo: O(n) Espacio: O(n) para la nueva lista
    try:
      return [a + b * x for x in list_estatura]
    except TypeError:
      logger.error("Los parámetros a, b y estatura deben ser numéricos.")
      raise
````

Se traduce el código utilizando terminología técnica precisa de Data Science (e.g., Ordinary Least Squares, intercept, slope) y ajustando rigurosamente la longitud de las líneas para cumplir con PEP 8 (máximo 79-80 caracteres).

````python
import logging
from typing import List, Tuple

# Logger basic configuration
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger("AI_Library")


def linear_regression(X: List[float], Y: List[float]) -> Tuple[float, float]:
    """
    Calculates the linear regression coefficients (y = a + bx).

    Uses the Ordinary Least Squares (OLS) method to find the slope (b)
    and the intercept (a) that best fit the data.

    Args:
        X: List of independent variables (e.g., height).
        Y: List of dependent variables (e.g., weight).

    Returns:
        A tuple (a, b) where 'a' is the intercept and 'b' is the slope.

    Complexity:
        Time: O(n) | Space: O(1)
    """
    logger.info("Starting linear regression calculation.")

    try:
        if len(X) != len(Y):
            raise ValueError("Arrays X and Y must have the same length.")

        n = len(X)
        if n == 0:
            raise ValueError("Input lists cannot be empty.")

        sum_x = sum(X)
        sum_y = sum(Y)
        sum_x2 = sum(x**2 for x in X)
        sum_xy = sum(x*y for x, y in zip(X, Y))

        denominator = (n * sum_x2) - (sum_x ** 2)

        if denominator == 0:
            raise ZeroDivisionError("X variance is zero (vertical points).")

        # Slope calculation (b)
        # Formula: b = [n*Σ(xy) - Σx*Σy] / [n*Σ(x²) - (Σx)²]
        b = ((n * sum_xy) - (sum_x * sum_y)) / denominator

        print("--- Calculation Process ---")
        print(f"b = ({n}*{sum_xy} - {sum_x}*{sum_y}) / "
              f"({n}*{sum_x2} - ({sum_x}^2))")
        print(f"b = {b}")

        a = (sum_y / n) - b * (sum_x / n)

        print(f"a = y_mean - b * x_mean")
        print(f"a = {Y} - {b}*{X}")
        print(f"a = {a}")
        print("---------------------------")

        logger.info(f"Model trained successfully: a={a:.4f}, b={b:.4f}")
        return a, b

    except ValueError as ve:
        logger.error(f"Validation Error: {ve}")
        raise
    except ZeroDivisionError as zde:
        logger.error(f"Math Error: {zde}")
        raise
    except Exception as e:
        logger.critical(f"Unexpected error in linear_regression: {e}")
        raise


def estimate_value(a: float, b: float, x_input: float) -> float:
    """
    Predicts a single value using the model (y = a + bx).

    >>> estimate_value(10.0, 2.0, 70.0)
    150.0
    """
    try:
        return a + b * x_input
    except TypeError:
        logger.error("Parameters a, b, and x_input must be numeric.")
        raise


def bulk_estimate(
    a: float,
    b: float,
    x_list: List[float]
) -> List[float]:
    """
    Performs predictions for a full list of values.

    Args:
        a: Y-intercept.
        b: Line slope.
        x_list: List of X input values.

    Returns:
        List of calculated estimates.
    """
    try:
        return [a + b * x for x in x_list]
    except TypeError:
        logger.error("Parameters a, b, and x_list must contain numeric types.")
        raise
````

#Optimización, ese es el "ojo clínico" de un Senior Developer.

Cuando hablamos de Optimización en Python para Ciencia de Datos, nos referimos a dos niveles: la eficiencia algorítmica y la eficiencia de implementación.

En una empresa internacional como amazon, google, loreal, etc., donde podrías procesar millones de transacciones, la diferencia entre un bucle for y una operación vectorizada es la diferencia entre minutos y milisegundos.

Aquí tienes los 3 tips de optimización que llevarán el código al siguiente nivel:

**1. Vectorización (El estándar de la industria)**
En lugar de procesar elemento por elemento, un experto usa NumPy. NumPy delega los cálculos a bibliotecas escritas en C y Fortran, realizando operaciones en paralelo.

Tip Senior: Si la lista tiene más de 1,000 elementos, deja de usar Python puro y usa arreglos

````python
import numpy as np

def optimized_regression(X: List[float], Y: List[float]):
    # Convertimos a arreglos de NumPy para cálculos vectorizados
    x_arr = np.array(X)
    y_arr = np.array(Y)

    # Las operaciones se aplican a todo el vector simultáneamente
    # Esto es mucho más rápido que un bucle for
    n = len(x_arr)
    sum_x = np.sum(x_arr)
    sum_xy = np.sum(x_arr * y_arr) # Multiplicación elemento a elemento
````

**2. Evitar la Redundancia de Cálculos**

En el código anterior, calculábamos sum_x / n varias veces. Un Senior almacena ese valor en una variable (como x_mean) para ahorrar ciclos de CPU.

**3. Uso de Generators para Memoria**

Si no puedes usar NumPy y los datos son masivos, usar Generadores (como hicimos con sum(x**2 for x in X)) es mejor que usar List Comprehensions sum([x**2 for x in X]).

* List Comprehension: Crea una lista completa en memoria antes de sumarla.

* Generator: Procesa un número a la vez y lo suma, manteniendo el uso de memoria en $O(1)$.

**Versión Optimizada Final (Pythonic & Efficient)**


````python
def linear_regression(X: List[float], Y: List[float]) -> Tuple[float, float]:
    """
    Optimized Linear Regression calculation.
    """
    n = len(X)
    # Using means to simplify the slope (b) formula
    x_mean = sum(X) / n
    y_mean = sum(Y) / n

    # Numerator and denominator for 'b' using generators for memory efficiency
    numerator = sum((X[i] - x_mean) * (Y[i] - y_mean) for i in range(n))
    denominator = sum((X[i] - x_mean) ** 2 for i in range(n))

    if denominator == 0:
        raise ZeroDivisionError("X variance is zero.")

    b = numerator / denominator
    a = y_mean - (b * x_mean)

    return a, b
````

**¿Por qué esta versión es "más Senior"?**

Esta fórmula (basada en desviaciones de la media) es matemáticamente equivalente a la anterior, pero es numéricamente más estable. Evita manejar números extremadamente grandes en las sumatorias de cuadrados, lo que previene errores de precisión en coma flotante (floating point errors).

#Medición de tiempo y espacio

En ingeniería de software de alto nivel, no basta con que el código funcione; debemos medir su desempeño. A esto le llamamos **Profiling**.

De forma profesional, no debemos "ensuciar" la función principal con cálculos de tiempo. Lo ideal es usar Decoradores o funciones auxiliares de medición.

Aquí tienes la implementación con dos funciones de monitoreo y la versión final de linear_regression que integra todo (Inglés, PEP8, Logging, Excepciones y Medición):

**1. Funciones de Monitoreo (Profiling Tools)**

````python
import time
import sys
from typing import Any, Callable

def trace_execution_time(func: Callable, *args: Any, **kwargs: Any) -> Any:
    """
    Measures the execution time of a function in seconds.
    """
    start_time = time.perf_counter()
    result = func(*args, **kwargs)
    end_time = time.perf_counter()

    duration = end_time - start_time
    # Usamos print para visibilidad inmediata en Colab/Notebooks
    print(f"[PROFILER] Execution Time: {duration:.8f} seconds")
    return result

def trace_memory_usage(name: str, data: Any) -> None:
    """
    Calculates the approximate memory storage of an object in bytes.
    """
    # sys.getsizeof + size of elements if it's a list
    size_bytes = sys.getsizeof(data)
    if isinstance(data, list):
        size_bytes += sum(sys.getsizeof(item) for item in data)

    print(f"[PROFILER] Storage Usage ({name}): {size_bytes} bytes")
````


**Claridad de Etiquetas**: El prefijo [PROFILER] para que se distinga qué es un mensaje de la lógica del programa y qué es una métrica de rendimiento.

Un **tip** extra:
Si quieren llevar esto a un nivel de "Senior", el Storage que estamos midiendo es en la RAM. Si el archivo viniera de un .csv pesado, el tiempo de ejecución sería mayor debido a la latencia de I/O (entrada/salida).

**2. Código integrado y profesional (Final versión)**



**Ubicación de la llamada:** se ha movido trace_memory_usage dentro de la función y trace_execution_time como un wrapper del llamado principal. Así, verás tres líneas de métricas antes del resultado final.


````python
import logging
import time
import sys
from typing import List, Tuple, Any

# Professional Logging Setup
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger("AI_Library")

def linear_regression(X: List[float], Y: List[float]) -> Tuple[float, float]:
    """
    Calculates OLS Linear Regression coefficients with
    performance monitoring.
    """
    # Tracing storage before processing
    trace_memory_usage("Variable X", X)
    trace_memory_usage("Variable Y", Y)

    logger.info("Starting linear_regression calculation...")

    try:
        if len(X) != len(Y) or len(X) == 0:
            raise ValueError("Invalid input dimensions.")

        n = len(X)

        # Optimization: Use means to avoid large sums (Numerical Stability)
        x_mean = sum(X) / n
        y_mean = sum(Y) / n

        # Numerator and Denominator using memory-efficient generators
        num = sum((X[i] - x_mean) * (Y[i] - y_mean) for i in range(n))
        den = sum((X[i] - x_mean) ** 2 for i in range(n))

        if den == 0:
            raise ZeroDivisionError("Vertical line detected (zero variance).")

        b = num / den
        a = y_mean - (b * x_mean)

        return a, b

    except Exception as e:
        logger.error(f"Failed to calculate regression: {e}")
        raise

# --- EXAMPLE OF HOW TO INVOKE (The Senior Way) ---

if __name__ == "__main__":
    # Sample Data
    heights = [1.60, 1.65, 1.70, 1.75, 1.80, 1.85, 1.90]
    weights = [60.5, 63.0, 70.2, 75.4, 82.1, 85.0, 90.5]

    try:
        # Wrap the call with the execution time tracer
        intercept, slope = trace_execution_time(
            linear_regression, heights, weights
        )

        print(f"\nModel: y = {intercept:.4f} + {slope:.4f}x")

    except Exception:
        print("Model training failed. Check logs.")
````

**¿Por qué este enfoque es nivel Senior?**

**Separación de Preocupaciones (Separation of Concerns)**:  La lógica matemática de la regresión está separada de la lógica de medición. Esto permite que si mañana quieres dejar de medir el tiempo, no tienes que tocar la función linear_regression.

**time.perf_counter():** Un novato usa time.time(). Un profesional usa perf_counter() porque tiene mayor resolución y no se ve afectado por actualizaciones del reloj del sistema.

**Análisis de almacenamiento:** La función trace_memory_usage ayuda a los Cienticos de Datos a entender cuánto pesan sus datasets en la RAM, algo crítico cuando se trabaja con Big Data.

**Estilo Limpio:** Todo se mantiene bajo los 80 caracteres de ancho, lo que facilita la lectura en pantallas divididas (Side-by-side) durante un Code Review.

3. El misterio de los __init__.py: **¿Por qué tantos?**
En Python, una carpeta no es automáticamente un "paquete". El archivo __init__.py es el "DNI" de la carpeta.

**¿Es necesario en todas?** Sí, si quieres que sean sub-paquetes.

La analogía: Imagina que unam_ai es un edificio. Sin __init__.py, Python ve el edificio como una caja cerrada. Con el archivo en cada piso (models/, utils/), Python puede entrar a cada habitación y sacar las funciones.

Importaciones limpias: Probar hacer  from unam_ai.models import linear_regression en lugar de rutas de archivo de sistema largas.

4. Cómo conviven los Notebooks con los archivos .py
Este es el dilema de muchos científicos de datos. La mejor forma de trabajar nivel Senior es:

**El código "duro" (matemáticas y lógica)**: Vive en src/unam_ai/. Esto es lo que se sube a GitHub como biblioteca.

**El Notebook (.ipynb)**: Se usa como un "laboratorio". En la primera celda del Notebook en Colab, el alumno instala su propia librería o la importa de la carpeta src.

5. Beneficios de este modelo de carpetas (El "Por qué" Senior)
* **Para Classroom**: El alumno entrega el .ipynb. Tú ves su explicación, sus gráficas y sus conclusiones, pero el código real viene de un motor bien estructurado en src.

* **Para GitHub**: El repositorio se ve profesional. Tiene una carpeta de código fuente y una de ejemplos/ejercicios.

**¿Por qué esta estructura de carpetas es la definitiva?**

1. **Home.py + pages/**: Streamlit reconoce automáticamente cualquier archivo dentro de pages/ como una pestaña nueva en el menú lateral. El orden se da por los números (01, 02...).

# Estrategia de Ambientes: DEV → STAGE → PROD


Para alumnos de Ciencia de Datos, es vital entender que el código no solo vive en su laptop. Aquí está la propuesta de ciclo de vida para este proyecto para esta materia y el proyecto que se tiene que entregar:

**Ambiente 1: DEV (Desarrollo)**

**Herramienta**: VS Code + Notebooks locales.

**Objetivo**: Escribir la lógica en src/ y probarla rápidamente en archivos .ipynb.

**Estado**: Es lo que estás haciendo ahora. El código cambia constantemente.

**Ambiente 2: STAGE (Pruebas/Pre-producción)**

**Herramienta**: GitHub Actions + tests/.

**Propuesta**: Configurar un sistema que, cada vez que el alumno suba código a GitHub, ejecute automáticamente el archivo tests/test_regression.py.

***Regla de ORO***: Si el test falla , el código no "pasa" a la siguiente etapa. Esto les enseña ***Integración continua*** (CI).

**Ambiente 3**: PROD (Producción)
Herramienta: Streamlit Cloud o Docker.

**Propuesta**: Desplegar la carpeta app_Streamlit en la web.

**Diferencia clave**: En PROD, ya no usamos sys.path.append. Instalamos la librería formalmente usando el archivo requirements.txt que creamos.

**Resultado**: Un link público (ej. unam-ia-lab.streamlit.app) donde cualquiera puede usar su modelo.

Recuerden que:

* Notebooks son para explicar y experimentar (el "por qué" de las cosas).

* Scripts .py son para ejecutar y producir (el "cómo" funciona el sistema).

Esa es la mentalidad de un **Arquitecto de Soluciones**. Mantener los tres ambientes (DEV, STAGE, PROD) sincronizados en un mismo repositorio de GitHub es lo que diferencia a un programador de un Ingeniero de Software.
