# Uso de librerías para arreglos y matrices en Numpy

La librería numérica de Python más conocida por su nombre simplificado es Numpy, una de las más relevantes dentro de este lenguaje, convirtiéndose en un estándar dentro de la comunidad de programación científica y en casi todos los desarrollos relacionados a la ciencia de datos y la inteligencia artificial.

Según McKinney (2017), algunos de los elementos claves de numpy son:

La implementación de un eficiente arreglo multidimensional (ndarray), el cual está orientado a optimizar la ejecución de operaciones aritméticas.
Contiene varias funciones matemáticas diseñadas para realizar operaciones con arreglos de datos significativamente grandes de forma rápida y mediante la escritura de un código más limpio.
Posee un conjunto de herramientas para la lectura y escritura de datos en almacenamientos externos utilizando ficheros mapeados en memoria.
Está optimizada para resolver problemas de álgebra lineal, la generación de valores aleatorios y realizar operaciones en el dominio de la frecuencia.
Implementa una interfaz de punto de acceso (API) en lenguaje C, que permite conectarla con librerías escritas en otros lenguajes de programación.
Numpy es muy utilizado para servir de interfaz entre antiguos programas escritos en C, C++ o Fortran y las nuevas formas de programación orientadas a arreglos; esta facilidad incrementa la vida útil de muchas aplicaciones científicas que de otra forma ya estarían descontinuadas. Es importante señalar que numpy en sí, no provee de muchas funciones para generar modelos matemáticos complejos, pero la capacidad de trabajo con arreglos y su forma específica de programación, se combinan de manera excelente con otras librerías científicas, creando una simbiosis imposible de superar.

Según McKinney (2017), para la mayoría de las aplicaciones de análisis de datos, las principales funcionalidades que esta librería ofrece son:

• Realización de complejas operaciones con matrices, de forma rápida y enfocadas en el intercambio, limpieza de datos, generación de subconjuntos, filtrado de valores, transformaciones entre tipos de datos o cualquier otro tipo de cálculo.

• Uso de algoritmos básicos para realizar operaciones dentro de las matrices; por ejemplo, la ordenación de elementos de forma individual o de conjuntos.

• Eficiente descripción de estadísticas, agregación o resumen de datos.

• Facilidad de manipular datos relacionales y realizar las operaciones unión o fusión de estos en nuevos conjuntos de datos.

• Utilizar elementos de lógica condicional como expresiones dentro de los arreglos, en lugar de utilizar bucles u otras estructuras de control.

• Manipulaciones de datos grupales (agregación, transformación o aplicación de funciones).


Una de las razones por las que numpy es tan importante para la computación numérica en Python, es porque está diseñado especialmente para trabajar con grandes arreglos de datos. Algunas de las consideraciones que se tuvieron a la hora de diseñar esta librería fueron (McKinney, 2017):

Almacenar los datos de forma interna en bloques contiguos de memoria, haciéndolo de forma diferente a como Python manejaría de forma nativa datos con iguales características, esto es posible por la forma en que opera el algoritmo de bajo nivel escrito en C y por supuesto agregando la ventaja de que se consume una menor cantidad de recursos de memoria.

Diseñar todas las operaciones computacionales con arreglos que realiza la librería, utilizando la menor cantidad posible de ciclos o estructuras de control.

**Arreglos en Numpy**

Una de las características principales de numpy es su arreglo n-dimensional, más conocido como “ndarray”, el cual es un contenedor rápido y flexible para cualquier conjunto de datos en Python. Un arreglo de numpy, nos permite realizar operaciones matemáticas a grandes bloques de datos, con la misma sencillez que se efectuaría con cada elemento de forma independiente.

Un “ndarray” es contenedor genérico multidimensional para datos homogéneos, lo cual significa que todos sus elementos tienen que ser del mismo tipo. Cada contenedor tiene una forma y un tipo de dato, esta información nos devuelve el tamaño y el tipo de valor que está alojando el arreglo durante su uso.

Un “ndarray” se crea de la siguiente forma

# Librerías para graficado en Matplotlib

Con la librería Matplotlib se pueden generar gráficos como líneas y diagramas de dispersión de una forma muy simple. En los siguientes ejemplos aprenderemos a utilizarla dentro de nuestro programa para visualizar primeramente un pequeño conjunto de datos definidos de forma manual y a continuación un conjunto más grande, generado automáticamente mediante una variable aleatoria.

Como primer ejemplo, Matthes (2019) propone trazar un gráfico de una línea simple empleando Matplotlib y después más adelante podrás personalizarlo para crear una visualización de datos con más información. Se utiliza la secuencia del valor de los números 1, 2, 3, 4 y 5 elevados al cuadrado como datos para el gráfico.

```python
#Gráfico de una secuencia de números usando Matplotlib.
import matplotlib.pyplot as plt
cuadrados = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()
ax.plot(cuadrados)
plt.show()
```

El primer paso fue importar el módulo pyplot usando el alias plt para no repetir toda la palabra, esa suele ser una práctica muy común al utilizarlo. El módulo contiene una serie de funciones que generan gráficos y diagramas.

Después, se crea una lista llamada cuadrados para contener los datos que se trazarán y entonces en la línea 3 se llama a la función subplots() de Matplotlib. Esta función puede generar uno o más gráficos en la misma figura.

La variable fig representa la figura completa o la colección de gráficos que se generen. La variable ax, por su parte, representa un solo gráfico en la figura y se usará la mayor parte del tiempo.  

Como últimos pasos para este ejemplo se usa el método plot(), cuya función es graficar los datos proporcionados, mientras que la función plt.show () abre el visor de Matplotlib y muestra el gráfico, como pudo apreciarse en el resultado de la Tabla 1. Sin embargo, al analizar el gráfico es que el valor de 4 en el eje horizontal no coincide con el valor de 25 en el eje vertical, por lo que en el siguiente ejemplo se ajustará y corregirá el gráfico agregando otra lista al programa llamada valores_entrada.

Aunque el gráfico anterior muestra el comportamiento de los datos, el tamaño de las etiquetas es demasiado pequeño y la línea un poco delgada para apreciarse fácilmente. Afortunadamente, Matplotlib permite ajustar cada característica del gráfico. A continuación, se usarán algunas de las personalizaciones disponibles para mejorar la legibilidad a través del siguiente código:

```python
#Gráfico de una secuencia de números ajustando parámetros.
import matplotlib.pyplot as plt
valores_entrada = [1, 2, 3, 4, 5]
cuadrados = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()
ax.plot(valores_entrada, cuadrados, linewidth=3)
#Establecer el título del gráfico y los ejes de la etiqueta.
ax.set_title("Números elevados al cuadrado", fontsize=14)
ax.set_xlabel("Base", fontsize=14)
ax.set_ylabel("Valor al cuadrado", fontsize=14)
#Establecer el tamaño de las etiquetas de las marcas.
ax.tick_params(axis='both', labelsize=14)
plt.show()
```

El parámetro linewidth utilizado en la línea 5 del código controla el grosor de la línea del gráfico generada por la función plot(); también se incluye a la lista valores_entrada como parámetro, declarada en la línea 2 del código, con el propósito de ajustar los valores del eje horizontal y coincidan con relación a los del eje vertical. El método set_title () en la línea 7 establece un título para el gráfico. Los parámetros de tamaño de fuente, que aparecen repetidamente en el código, controlan el tamaño del texto en los elementos del gráfico.

Por su parte, los métodos set_xlabel() y set_ylabel() declarados en la línea 8 permiten establecer un título para cada uno de los ejes x y y, y el método tick_params() de la línea 11 estiliza las marcas de graduación, sus argumentos afectan las marcas de graduación eje x y y (axis='both') y establece el tamaño de fuente de las etiquetas en 14 (labelsize=14).

Como puedes ver a continaución, resulta mucho más fácil leer el gráfico en comparación con el generado por el código de la gráfica anterior, ya que el tipo de etiqueta es más grande y el gráfico de líneas es más grueso. A menudo, vale la pena experimentar con los diferentes parámetros para encontrar cual es el mejor gráfico resultante.

Sin embargo, escribir listas a mano puede resultar ineficaz, sobre todo cuando se tienen muchos puntos. Por lo que ahora, en vez de pasar los puntos a una lista, se usará un bucle para que Python los genere automáticamente.

Si se desea guardar el gráfico en un archivo se puede hacer por medio de la función plt.savefig().

El siguiente ejemplo sirve para 1000 puntos:

```python
#Gráfico de una secuencia de números generada automáticamente.
import matplotlib.pyplot as plt
x_valores = range(1, 1001)
y_valores = [x**2 for x in x_valores]
plt.style.use('seaborn-v0_8')
#prueba con este otro estilo y usa el que más te guste
#plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.scatter(x_valores, y_valores, s=10)
#Establecer el título del gráfico y los ejes de la etiqueta.
ax.set_title("Números elevados al cuadrado", fontsize=14)
ax.set_xlabel("Base", fontsize=14)
ax.set_ylabel("Valor al cuadrado", fontsize=14)
#Establezca el rango para cada eje.
ax.axis([0, 1100, 0, 1100000])
plt.savefig('grafico_cuadrados.png', bbox_inches='tight')
plt.show()
```

A continuación, se analizarán los resultados de tirar un dado. Cuando se lanza un dado normal de seis caras, cualquiera de los números del 1 al 6 tienen la misma probabilidad de salir. Entonces, se representará dicha probabilidad por medio de un conjunto de datos. Finalmente, se trazará un histograma usando Plotly.

Para hacer uso de Plotly en Python es necesario instalarlo previamente.


Una vez descargado Plotly, puedes comenzar realizando el siguiente programa para crear una clase llamada Dado, con la cual se puede crear un dado de seis lados y lanzarlo por medio del método lanzar():

```python
#Declaración de la clase Dado.
from random import randint
class Dado:
  """Clase que representa a un dado."""
  def __init__(self, num_lados=6):
    """Se consideran a un dado con 6 lados."""
    self.num_lados = num_lados
  def lanzar(self):
    """"Devuelve un valor aleatorio entre 1 y el número de lados."""
    return randint(1, self.num_lados)
```

Antes de crear una visualización basada en la clase declarada anteriormente, se crea el objeto dado y se imprimen los resultados para verificar que los resultados sean razonables:

```python
#Se crea un dado de seis lados
#Simulación de múltiples lanzamientos de un dado.
dado = Dado()
#Se realizan algunos lanzamientos y se guardan en una lista.
resultados = []
for num_lanzamiento in range(100):
  resultado = dado.lanzar()
  resultados.append(resultado)
print(resultados)
```

Un análisis rápido de los resultados anteriores permite ver que la clase Dado funciona. Se aprecian valores entre 1 y 6, por lo que la función está devolviendo los valores más pequeños y más grandes posibles, y como no sale ningún 0, 7 u otra cosa, se ha comprobado que todos los resultados están en el rango apropiado.

Mediante el último ejemplo se determinará cuántas veces aparece cada número exactamente. Con una lista de frecuencias, es posible hacer un histograma de los resultados, el cual es un gráfico de barras que muestra la frecuencia con la que ocurren ciertos resultados. El código para crearlo sería el siguiente:

```python
from plotly.graph_objs import Bar, Layout
from plotly import offline
#Análisis de los resultados.
frecuencias = []
for valor in range(1, dado.num_lados+1):
  frecuencia = resultados.count(valor)
  frecuencias.append(frecuencia)

#Visualización de los resultados
x_valores = list(range(1, dado.num_lados+1))
datos = [Bar(x=x_valores, y=frecuencias)]

x_eje_config = {'title': 'Resultado'}
y_eje_config = {'title': 'Frecuencia del resultado'}
my_layout = Layout(title='Resultados de lanzar el dado 1000 veces',

xaxis=x_eje_config, yaxis=y_eje_config)

offline.plot({'data': datos, 'layout': my_layout}, filename='Dado.html')
```

```python
import matplotlib.pyplot as plt

# --- OPERACIÓN 1: Seleccionar una muestra y organizar ---
# Tomamos una muestra de 500 filas para que el gráfico no sea un caos
muestra = df_mexico.sample(n=50, random_state=42)

# --- OPERACIÓN 2: Aplicar una función matemática ---
# Digamos que queremos calcular un índice simple de 'Carga de Partículas'
# multiplicando PM2.5 por un factor de riesgo (ejemplo hipotético)
muestra['carga_riesgo'] = muestra['PM2.5'] * 1.5

# --- OPERACIÓN 3: Representación Gráfica ---
plt.style.use('seaborn-v0_8') # Elige el que más te guste
fig, ax = plt.subplots(figsize=(10, 6))

# Graficamos la muestra
ax.scatter(muestra['PM2.5'], muestra['carga_riesgo'], c='orange', alpha=0.6)

ax.set_title("Análisis de Riesgo: Carga de Partículas PM2.5", fontsize=14)
ax.set_xlabel("Nivel de PM2.5", fontsize=12)
ax.set_ylabel("Índice de Riesgo Calculado", fontsize=12)

# ORDEN CORRECTO: Guardar y luego mostrar
plt.savefig('resultado_calidad_aire.png', bbox_inches='tight')
plt.show()

print("¡Tarea completada! El archivo 'resultado_calidad_aire.png' ha sido generado.")
```
